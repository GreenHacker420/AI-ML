```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler # For numerical feature standardization
```

```python
# Read dataset
df = pd.read_csv('ques_data.csv')
```

```python
print("Initial Head of the DataFrame:")
df.head()
```

```python
print("\nInitial Info of the DataFrame:")
df.info()
```

```python
# removing not useful columns

df.drop('CustomerID',axis=1,inplace = True)
```

```python
print("\nInitial Missing Values Count:")
df.isnull().sum()
```

```python
print("\nInitial Duplicate Rows Count:")
df.duplicated().sum()
```

```python
# ---  Data Cleaning - Duplicates  --- 


#write code below to drop duplicate values

df.drop_duploicates(inplace=True)


df.reset_index(drop=True, inplace=True)
rows_after_duplicatesRemoved = df.shape[0]
```

```python
rows_after_duplicatesRemoved
```

```python
# --- Missing Value Handling ---
print("\n--- Task 3: Missing Value Handling ---")

# Numerical columns for imputation
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
# Exclude the target variable if it's numerical and doesn't have missing values
if 'Has_Subscription' in numerical_cols:
    numerical_cols.remove('Has_Subscription')

# Categorical columns for imputation
categorical_cols = df.select_dtypes(include='object').columns.tolist()

print(f"Numerical columns for imputation: {numerical_cols}")
print(f"Categorical columns for imputation: {categorical_cols}")

# Impute numerical columns with median
for col in numerical_cols:
    median_value = df[col].median()
    df[col].fillna(median_value, inplace=True)

# write code here

# Impute categorical columns with mode
for col in categorical_cols:
    mode_value = df[col].mode()[0]
    df[col].fillna(mode_value, inplace=True)

# write code here
```

```python
df.isnull().sum().to_dict()
```

```python
print(f"Age max value{max(df['Age'])}")
print(f"Age min value{min(df['Age'])}")
print(f"Monthly_Spend max value{max(df['Monthly_Spend'])}")
print(f"Monthly_Spend min value{min(df['Monthly_Spend'])}")
```

```python
# --- Outlier Handling (for Age & Monthly_Spend) ---

# Columns to perform outlier handling on

outlier_columns = ['Age', 'Monthly_Spend']

for col in outlier_columns:
    if col in df.columns :
        
        #write code to find Q1
        q1 = df[col].quantile(0.25)
        
        #write code to find Q3
        q3 = df[col].quantile(0.75)
        # write code to Find IQR , lower_bound, upper_bound, than cap the values for upper and lower
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        df[col] = np.where(df[col] < lower_bound, lower_bound, df[col])
        df[col] = np.where(df[col] > upper_bound, upper_bound, df[col])
        # Ensure no negative values after flooring (especially relevant for Monthly_Spend, but good for Age too)
        df[col] = np.maximum(0, df[col])
        

    else:
        print(f"Outlier handling skipped for '{col}': not found or not numeric.")

```

```python
# ---  Encoding Categorical Features ---

# Apply One-Hot Encoding to 'Product_Category'
df = pd.get_dummies(df, columns=['Product_Category'], drop_first=True)


```

```python
# ---  Perform Standardization for Numerical Features ---

# Identify numerical features to scale (all numeric except the target variable)
features_to_standardize = ['Age','Monthly_Spend']
print(f"Numerical features to standardize: {features_to_standardize}")


# perform standard scaling here
scale = StandardScaler()
df[features_to_standardize] = scale.fit_transform(df[features_to_standardize])

```
